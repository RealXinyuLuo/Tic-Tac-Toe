```python
from IPython.display import clear_output as clear
import sys
```

```python
## Game setup 
class State:        
    element_row1 = [' ',' ',' '];
    element_row2 = [' ',' ',' '];
    element_row3 = [' ',' ',' '];
    element_matrix = [element_row1,element_row2,element_row3];

    to_print = [];

    game_on = True;
    input_check = False;
    player_flag = 1;
    has_winner = False
    
    # Defalt names for players 1 and 2
    player1 = "playerA";
    player2 = "playerB";
    winner = "";
    
    def __init__(self):
        
        self.player1 = input("Hello, player one. Please enter your name:")
        self.player2 = input("Hello, player two. Please enter your name:")
        
        print("Welcome to the game, {} and {}.".format(self.player1,self.player2))
        print("{}, you will be using 'x'".format(self.player1))
        print("{}, you will be using 'o'".format(self.player2))
        
    
def setup():
    _state = State();
    #print("game has initiated")
    return _state  
```

```python
## Game input 
def game_input(_state):
    """
    takes player input for coords and run validate() to check
    --------------------
    input: user input 
    output: 2D x-y coordinate as a tuple 
    """
    x_flag = False; #default flag value so game can start 
    y_flag = False; #default flag value so game can start
    availability= True #default flag value so game can start
    
    while availability:
        while not x_flag:
            if _state.player_flag == 1:
                x = input("{}, please enter the x coordinate:".format(_state.player1)); 
            else:
                x = input("{}, please enter the x coordinate:".format(_state.player2));          
            
            if validate(x):
                x_flag = True;    #if value valid change default x flag to stop while loop
                x_coord = int(x);

        while not y_flag:
            if _state.player_flag == 1:
                y = input("{}, please enter the y coordinate:".format(_state.player1));
            else:
                y = input("{}, please enter the y coordinate:".format(_state.player2));
            
            if validate(y):
                y_flag = True;    #if value valid change default y flag to stop while loop
                y_coord = int(y);
        
        #check if the selected position is still available 
        if _state.element_matrix[x_coord][y_coord] == ' ':
            availability = True
            return (x_coord,y_coord)
        else:
            availability_flag = False
            x_flag = False
            y_flag = False
            print("This board position is already occupied! Please choose another one! ") 
```

```python
## Validate user input 
def validate(_input):
    """
    check if the input is an int between 0 and 2
    --------------------
    input: value to be check
    output: True if value is an int between 0 and 2
            False otherwise and print error message 
    """
    if _input.isdigit() and int(_input) in range(0,3):
        return True
    elif not _input.isdigit():
        print("Please enter an integer!")
        return False
    elif _input.isdigit() and int(_input) not in range(0,3):
        print("Please enter an integer between 0 and 2!")
        return False
```

```python
## Player switch 
def player_switch(_state):
    """
    If player one, switch to player two.
    If player two, switch to player one. 
    """
    if _state.player_flag == 1:
        _state.player_flag = 2;
    elif _state.player_flag == 2:
        _state.player_flag = 1;
    else: 
        print("Player flag error: flag is not 1 or 2.")
        sys.exit();
```

```python

## Evaluate win condition
def who(_element,_state):
    """
    Return the name of the winner. 
    """
    if _element == 'x':
        return _state.player1
    elif _element == 'o':
        return _state.player2 
    
def who_wins(_state):
    """
    Win check logic abstraction.
    Update the winner information. 
    Divided into three parts:
    1. row win 
    2. column win 
    3. diagnal win 
    """
    
    # 1.Check if any row has three identical elements   
    for row in _state.element_matrix:
        row_set = set(row)
        if len(row_set) == 1 and not row[0] == ' ':
            _state.game_on = False
            _state.winner = who(row[0],_state);
            _state.has_winner = True
        else:
            pass
            
    # 2.Check if any column has three identical elements 
    for i in range(len(_state.element_matrix)):
        if _state.element_matrix[0][i] == _state.element_matrix[1][i] and _state.element_matrix[1][i] == _state.element_matrix[2][i]:
            if not _state.element_matrix[0][i] == ' ':
                _state.game_on = False
                _state.winner = who(_state.element_matrix[0][i],_state)
                _state.has_winner = True 
        else:
            pass 
            
     # 3.Check if any diagnal has three identical elements 
    if _state.element_matrix[0][0] == _state.element_matrix[1][1] and _state.element_matrix[1][1] == _state.element_matrix[2][2]:
        if not _state.element_matrix[1][1] == ' ':
            _state.game_on = False
            _state.winner = who(_state.element_matrix[1][1],_state)
            _state.has_winner = True
    elif _state.element_matrix[0][2] == _state.element_matrix[1][1] and _state.element_matrix[1][1] == _state.element_matrix[2][0]:
        if not _state.element_matrix[1][1] == ' ':
            _state.game_on = False
            _state.winner = who(_state.element_matrix[1][1],_state)
            _state.has_winner = True 
    else:
        pass 

## Check if win condition has been met 
def check_win(_state):
    """
    Check win. Logic abstracted in who_wins().
    
    """
    who_wins(_state)     
    
    if _state.has_winner:
        print('Game has ended with {} as the winner!'.format(_state.winner))
    else:
        pass
        

## Player end game 
def user_end(_state):
    """
    Ask if the player wants to end game mannually before the win 
    condition is met. If Y, end game. If N continue.    
    """
    
    # Display different message depending on whose turn it is 
    if _state.player_flag == 1:
        end_player = input("{}, do you want to end the game now? Y/N".format(_state.player1))
    else:
        end_player = input("{}, do you want to end the game now? Y/N".format(_state.player2))
    
    
    
    while not (end_player == 'Y' or end_player == 'N'):
        print('end_player can only take values of "Y" and "N"!')
        end_player = input("End game now? Y/N ")

    if end_player == "Y":
        _state.game_on = False;
        print('Game has been ended by player.')
    elif end_player == "N":
        _state.game_on = True; 
        
### Endgame total 
def end_game(_state):
    """
    Change the game_on flag when the win condition is met.
    Return message congratulating the winner. 
    """

    check_win(_state);
    
    if _state.game_on:
        user_end(_state);    
```

```python
def play(_state):
    """
    Calls game_input to take user input 
    and put it on the game board 
    """  
    if _state.player_flag == 1:
        (x_coord,y_coord) = game_input(_state);
        _state.element_matrix[x_coord][y_coord] = 'x';
    elif _state.player_flag == 2:
        (x_coord,y_coord) = game_input(_state);
        _state.element_matrix[x_coord][y_coord] = 'o';     
    else: 
        print("Player flag error: flag other than 1 or 2 passed to play()")
        sys.exit();
```

```python
## Display game 
def display(_state):
    """
    Display the current state
    of the game to the player 
    """
    #Clear the previous round 
    clear()

    #Convert to print format
    row1 = '|'.join(_state.element_row1)
    row2 = '|'.join(_state.element_row2)
    row3 = '|'.join(_state.element_row3) 
    _state.to_print = [row1,row2,row3] 
    
    print(_state.to_print[0] + "\n" + _state.to_print[1] + "\n" + _state.to_print[2])
```

```python
## Run game and update state stats
def game(_state):
    #print("Game has started. {}, it's your turn".format(_state.player1));
    
    while _state.game_on:
        play(_state) 
        display(_state)
        end_game(_state)
        player_switch(_state) 
    return _state  #we might not need this return value?
```

```python
## Main 
game_state = setup();
game_state = game(game_state)
```
